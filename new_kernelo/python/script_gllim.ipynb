```python
Pi = np.ones(K) * 1 / K
A = np.ones((D, L, K))
C = np.ones((L, K)) * 2
Gamma = np.ones((L, L, K)) * 0.2
B = np.ones((D, K)) * 3
Sigma = np.ones((D, D, K)) * 0.3

gllim.setParamPi(Pi)
gllim.setParamA(A)
gllim.setParamC(C)
gllim.setParamGamma(Gamma)
gllim.setParamB(B)
gllim.setParamSigma(Sigma)

theta = gllim.getParams()
print(theta.Pi)
print(theta.A)
print(theta.C)
print(theta.Gamma)
print(theta.B)
print(theta.Sigma)
```

```python
new_theta = lib.GLLiMParameters(L,D,K)

Gamma = np.ones((L,L,K)) + np.random.rand(L,L,K) * 0.1
Sigma = np.ones((D,D,K)) + np.random.rand(D,D,K) * 0.1
for k in range(K):
    Gamma[:,:,k] += np.eye(L) * 0.1
    Gamma[:,:,k] = np.dot(Gamma[:,:,k], Gamma[:,:,k].T) * 0.001
    Sigma[:,:,k] += np.eye(D) * 0.1
    Sigma[:,:,k] = np.dot(Sigma[:,:,k], Sigma[:,:,k].T) * 0.001
new_theta.Pi = Pi
new_theta.A = A * 2
new_theta.C = C * 2
new_theta.Gamma = Gamma
new_theta.B = B * 2
new_theta.Sigma = Sigma

gllim.setParams(new_theta)

theta = gllim.getParams()
print(theta.Pi)
print(theta.A)
print(theta.C)
print(theta.Gamma)
print(theta.B)
print(theta.Sigma)
```
